```python
!pip install numpy==1.26 scipy==1.12 matplotlib==3.8
import numpy as np
import matplotlib.pyplot as plt


D = 1e-6
size = 50
time_steps = 100
tumor_area = np.zeros((size, size))


tumor_area[0,:] = 1.0
for t in range(time_steps):
    tumor_area[1:-1, 1:-1] += D * (
        tumor_area[2:, 1:-1] + tumor_area[:-2, 1:-1] +
        tumor_area[1:-1, 2:] + tumor_area[1:-1, :-2] - 4 * tumor_area[1:-1, 1:-1]
    )


plt.imshow(tumor_area, cmap='hot')
plt.colorbar(label='Drug Concentration')
plt.title('Simple Drug Diffusion Simulation')
plt.show()
```
